# Stock Images data Scrapping :

```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
```

```python
link='https://stock-pictures.netlify.app/'
```

```python
driver=webdriver.Chrome()
```

```python
driver.get(link)
```

```python
driver.page_source
```

```python
soup=BeautifulSoup(driver.page_source,'html.parser')
```

```python
data1=[]
for sp  in soup.find_all('div',class_='container'):
    link=sp.find('img').get('src')
    if 'jpg' in sp.find('img').get('src'):
        link=sp.find('img').get('src')
        tags=sp.find('div',class_='tags').text[7:].split(' ')
        likes=sp.find('div',class_='likes-comments').find_all('span')[0].text
        comments=sp.find('div',class_='likes-comments').find_all('span')[1].text
        data1.append([link,''.join(tags),likes,comments])
```

```python
import pandas as pd
df=pd.DataFrame(data1,columns=['link of image','Tags','Likes','Comments'])
```

```python
df.sort_values('Likes')
```

# Downloading the image :

#### getting the binary code from the link and storing it in the file to get the image.

```python
for link in  df['link of image']:
    res=requests.get(link)
    res.content
    fd=open('stockimages/'+link.split('/')[-1],'wb')
    fd.write(res.content)
    fd.close()

    
```
